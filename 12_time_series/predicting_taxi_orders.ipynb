## Обучение

### Создание признаков

Создадим признаки для горизонта прогнозирования в один шаг. Напишем функцию для создания признаков. Тип данных признака "num_orders" `datetime64` уже содержит нужную информацию, необходимо представить её как отдельные столбцы (календарные признаки): месяц, день, день недели и час.

Для вычисления отстающих значений добавим в функцию новый аргумент max_lag, который задаст максимальный размер отставания. Новые признаки назовем: 'lag_1', 'lag_2' — и до величины max_lag.

Вычислим скользящее среднее и добавим его как признак 'rolling_mean'. В функцию добавим новый аргумент rolling_mean_size, который задаст ширину окна.

```python
def make_features(data, max_lag, rolling_mean_size):
    data['difference'] = abs(data['num_orders'] - data['num_orders'].shift())
    data['month'] = data.index.month
    data['day'] = data.index.day
    data['dayofweek'] = data.index.dayofweek
    data['hour'] = data.index.hour
    
    for i in range(1, max_lag + 1):
        data['lag_' + str(i)] = data['num_orders'].shift(i)
    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()
    data['std'] = data['num_orders'].shift().rolling(rolling_mean_size).std()

make_features(data, 15, 15)
```

Разделим имеющийся датасет на 3 выборки: обучающую (для обучения модели), валидационную (для более качественной проверки обученной модели) и тестовую в соотношении 8:1:1 (80%:10%:10%). По условиям задачи тестовая выборка должна составлять 10% от исходных данных.

В качестве базовой обучим модель LinearRegression, а также еще несколько моделей, показавших наиболее хорошие результаты в прошлых проектах.

Для того, чтобы отмаштабировать признаки, воспользуемся одним из методов масштабирования — стандартизации данных.
Преобразуем обучающую,валидационную и тестовую выборки функцией transform().

```python
scaler = StandardScaler()
scaler.fit(features_train)

features_train = scaler.transform(features_train)
features_valid = scaler.transform(features_valid)
features_test = scaler.transform(features_test)
```

По условию задачи для оценки качества моделей необходимо применять метрику **RMSE**. Напишем функцию для расчета метрики RMSE:

```python
def rmse(target, predictions):
    mse = mean_squared_error(target, predictions)
    rmse = sqrt(mse)
    return rmse
```

### Модель LinearRegression 

```python
model_lr = LinearRegression()
```

```python
model_lr.fit(features_train, target_train)
predicted_lr = model_lr.predict(features_valid)
```

```python
print('RMSE LinearRegression на валидационной выборке: {:.3f}'.format(rmse(target_valid, predicted_lr)))
```

### Модель XGBRegressor

Напишем функцию подбора гиперпараметров. Для подбора гиперпараметров воспользуемся инструментом GridSearchCV.

```python
def grid(model, parameters, features, target):
    
    tscv = TimeSeriesSplit(n_splits=2)
    search = GridSearchCV(model, parameters, cv=tscv, n_jobs=-1)
    search.fit(features, target)
    
    print(search.best_params_)
```

```python
model_xgbr = XGBRegressor(random_state=12345, objective="reg:squarederror", verbosity=0)

xgbr_parameters = dict(n_estimators=range(50,251,50),
                       max_depth=range(2,8,2),
                       learning_rate=[0.1, 0.3, 0.5])
```

```python
%%time
grid(model_xgbr, xgbr_parameters, features_train, target_train)
```

```python
model_xgbr = XGBRegressor(random_state=12345, objective="reg:squarederror", n_estimators=250, max_depth=6, learning_rate=0.1, verbosity=0)
```

```python
predicted_xgbr = model_xgbr.predict(features_valid)
```

```python
print('RMSE XGBRegressor на валидационной выборке: {:.3f}'.format(rmse(target_valid, predicted_xgbr)))
```
